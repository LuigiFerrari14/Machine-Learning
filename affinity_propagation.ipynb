```python
# imports
import numpy as np
from sklearn.cluster import AffinityPropagation
```

```python
#loading data
X = np.array( [
[3, 4, 3, 2, 1],
[4, 3, 5, 1, 1],
[3, 5, 3, 3, 3],
[2, 1, 3, 3, 2],
[1, 1, 3, 2, 3]
])
```

```python
#Training

#model
model = AffinityPropagation()

#training
model.fit(X)

#fit - clustering
labels = model.predict(X)

#labels
labels
```

```python
# Another Exemple

#Imports
from sklearn import datasets as dt
import matplotlib.pyplot as plt

X, _ = dt.make_blobs( n_samples=300, 
                      centers=4,
                      cluster_std=0.60,
                      random_state=0 )
```

```python
plt.scatter(X[:,0], X[:,1], alpha=0.7, edgecolors='b')
```

```python
#Training

#model
model = AffinityPropagation(preference=-50)

#training
model.fit(X)

#fit - clustering
labels = model.predict(X)

#show clusters
fig = plt.figure
plt.title('Clustering')
plt.scatter(X[:,0], X[:,1], c=labels, cmap='rainbow', alpha=0.7, edgecolors='b')
```
